```python
import json 
import pandas as pd 
from pandas import json_normalize
```

```python
with open('all_data10.json') as f:
    d = json.load(f)

d['Deaths']
```

```python
# json_normalize(d['Deaths'])
deaths = pd.DataFrame()
# d['Deaths'].keys()
for key in d['Deaths'].keys():
    death_year = json_normalize(d['Deaths'][key])
    death_year = death_year.assign(year = key)
    deaths = pd.concat([deaths, death_year])

df = pd.DataFrame(deaths.set_index('year').stack(future_stack=True), columns=['deaths'])
```
